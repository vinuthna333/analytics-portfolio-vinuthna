```python
#Selecting only the rows where 'rating' contains 'mins' and targeting the 'rating' and 'delivery_time' columns ['rating','delivery_time']
#Then reassigning their values in the correct order ['delivery_time','rating']. 
#Column names remain the same only their values are reassigned.
data.loc[data['rating'].str.contains('mins'), ['rating','delivery_time']] = data.loc[data['rating'].str.contains('mins'), ['delivery_time','rating']].values
```

Check row 26. There are rows in 'delivery_time' column where the value is in a range '11-21 mins'. Taking the max (21 mins) to set realistic expectations.

```python
#1.Extracts both numbers if a range exists (e.g., "11-21 mins" → min_time=11, max_time=21).
#2.Handles single values correctly (e.g., "33 mins" → min_time=33, max_time=NaN).
#3.Fills NaN values in max_time with min_time (so "33 mins" is treated as min=33, max=33).
#(\d+)-?(\d+)? is used to extract numbers from a delivery_time string that may contain either a single number (e.g., "33 mins") or a range (e.g., "11-21 mins").
#he regex pattern (r'(\d+)-?(\d+)?') only extracts the numbers and does not include "mins" in the output. This is because "mins" is not inside the capturing groups () in the regex.
```

```python
data[['min_time', 'max_time']] = data['delivery_time'].str.extract(r'(\d+)-?(\d+)?')
data['max_time'].fillna(data['min_time'], inplace=True)
data['delivery_time'] = data['max_time']
```

Do Restaurants with Offers Have Better Ratings?

```python
plt.figure(figsize=(20, 10))
data["has_offer"] = data["offer"].apply(lambda x: "Yes" if x != "No Offer" else "No")
sns.boxplot(x="has_offer", y="rating", data=data, palette="Set2")
plt.title("Ratings Comparison: Restaurants with Offers vs. No Offers")
plt.xlabel("Has Offer?")
plt.ylabel("Rating")
plt.show()
```

Hypothesis: Do Restaurants with Offers Have Higher Ratings?

H₀ (Null Hypothesis): Restaurants with offers have the same average rating as those without offers.

H₁ (Alternative Hypothesis): Restaurants with offers have a higher average rating than those without.

T-Test (Comparing Two Groups)


```python
from scipy.stats import ttest_ind

# Split data into two groups: Restaurants with offers and without
with_offer = data[data["offer"] != "No Offer"]["rating"].dropna()
without_offer = data[data["offer"] == "No Offer"]["rating"].dropna()

# Perform an independent t-test
t_stat, p_value = ttest_ind(with_offer, without_offer, equal_var=False)  # Welch's T-test (handles unequal variances)

print(f"T-Statistic: {t_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

# Conclusion
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject H₀: Restaurants with offers have significantly different ratings.")
else:
    print("Fail to Reject H₀: No significant difference in ratings.")
```

Hypothesis: Does Location Affect Delivery Time?

H₀: The average delivery time is the same across different locations.

H₁: At least one location has a significantly different delivery time.

ANOVA (Comparing Multiple Groups)

```python
from scipy.stats import f_oneway

# Group delivery times by location
delivery_times = [group["delivery_time(mins)"].dropna() for _, group in data.groupby("location")]

# Perform ANOVA test
f_stat, p_value = f_oneway(*delivery_times)

print(f"F-Statistic: {f_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

if p_value < alpha:
    print("Reject H₀: Location has a significant effect on delivery time.")
else:
    print("Fail to Reject H₀: No significant difference in delivery time across locations.")
```

Hypothesis: Does Food Type Affect Rating?

H₀: The average rating is the same across different food types.

H₁: At least one food type has a significantly different rating.

ANOVA (Comparing Multiple Food Types)

```python
# Explode food types (since some restaurants serve multiple types)
food_types_expanded = data.assign(food_type=data["food_type"].str.split(", ")).explode("food_type")

# Group ratings by food type
ratings_by_food = [group["rating"].dropna() for _, group in food_types_expanded.groupby("food_type")]

# Perform ANOVA test
f_stat, p_value = f_oneway(*ratings_by_food)

print(f"F-Statistic: {f_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

if p_value < alpha:
    print("Reject H₀: Food type significantly affects restaurant ratings.")
else:
    print("Fail to Reject H₀: No significant difference in ratings across food types.")
```

Summary of Tests

If p-value < 0.05, we reject the null hypothesis and conclude that there is a significant difference.

If p-value ≥ 0.05, we fail to reject the null hypothesis, meaning no strong evidence for a difference.
